<h1>Data Cleaning</h1>

```python
df=pd.read_excel('./data/Customer Call List.xlsx')
df
```

<h1>Droping Duplicate values</h1>

```python
#null, format problem,/ ,... note usefull column , duplicate 
# drop duplicate
df.drop_duplicates() # remove the 20 row which is duplicated
```

<h1>Removeing unwanted column</h1>

```python
df=df.drop(columns='Not_Useful_Column')
```

<h1>Last name there is / ... _ and unwanted data </h1>

```python
df['Last_Name'].str.lstrip("...") # takes left and right  truncate left ....
```

```python
df['Last_Name'].str.lstrip("...")
```

<h1>Remove unwanted characters using strip regular expression normaly list all unwanted characters method</h1>

```python
df['Last_Name']=df['Last_Name'].str.strip("123._/") # truncate number,.,underscore and / slash
df
```

<h1>Phone number destored data</h1>

```python
#1. strip non used characters to seems like 123-545-5421	
#2.reformat to seem like above
#df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')# not conatin a-z or 0-9 replace empty
df['Phone_Number'].str.replace('/','')
df['Phone_Number'].str.replace('|','')
```

```python

df['Phone_Number']=df['Phone_Number'].str.strip("-_/|")
df['Phone_Number']=df['Phone_Number'].str.replace("nan",'')
df['Phone_Number']=df['Phone_Number'].str.replace("Nan",'')
df['Phone_Number']=df['Phone_Number'].str.replace("N/a",'')
df['Phone_Number']=df['Phone_Number'].str.replace("/",'')
df['Phone_Number']=df['Phone_Number'].str.replace("-",'')
df['Phone_Number']=df['Phone_Number'].str.replace("|",'')
df['Phone_Number']
```

```python
# to inser - just change all to tostr 
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
#use lambda to insert - after 3 index
df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df
```

```python
# split with comma b/c already separated and Expand help us to see in expanded table form


df[['Street Address','state','zip code']]=df['Address'].str.split(',',expand=True)


# now delated the old address and use the separated new 
```

```python
# cheking Paying customer y to yes N no 
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','n')
df
df
```

```python
df=df.replace('N/a','')

df
```

```python
df=df.fillna('')
df
```

<h1>Drop row if phone is missing or donot contact if value is y </h1>

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] =='': # check for each row x, and phone number is empty
        df.drop(x,inplace=True)
df
```

```python
# remove donot contact persons 
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] =='y': # check for each row x, and phone number is empty
        df.drop(x,inplace=True)
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

<h1> The index is 0 -3 not serious so confusing hence reset the index to retart </h1>

```python
df=df.reset_index(drop=True) # drop True will drop the previous and apply new starting from 0
df
```
